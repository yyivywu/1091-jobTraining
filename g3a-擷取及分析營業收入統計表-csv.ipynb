```python
# 輸入月營收年度及月份

year = input('請輸入民國年度（譬如:109）：')
month = input('請輸入月份（譬如:2）：')
```

```python
# 擷取資料

import pandas as pd

try:
    url = 'https://mops.twse.com.tw/nas/t21/sii/t21sc03_'+year+'_'+month+'.csv'
    #df = pd.read_csv(url)
    df = pd.read_csv('g3a-t21sc03_109_2.csv', encoding='utf-8-sig')
except Exception as e:
    print('**WARRN: 擷取失敗')
    print('錯誤訊息', e)
else:
    print('成功擷取!')
```

```python
# 觀看 DataFrame 相關資訊

#df.head(10)
#df.info()   # 欄位資訊
#df.columns  # 欄位名稱
df.dtypes   # 欄位型別
#df.shape    # 列數及欄數
#df.describe()   # 各項統計資訊
```

```python
# 轉換欄位的資料型別

df = df.astype({'公司代號': 'str'})
df.dtypes
```

```python
# 刪除欄資料及欄名稱更名

df.drop('出表日期', axis=1, inplace=True)   # 刪除

df.rename(columns={'營業收入-當月營收':'當月營收'}, inplace=True)
df.rename(columns={'營業收入-上月營收':'上月營收'}, inplace=True)
df.rename(columns={'營業收入-去年當月營收':'去年當月營收'}, inplace=True)
df.rename(columns={'營業收入-上月比較增減(%)':'上月比較增減'}, inplace=True)
df.rename(columns={'營業收入-去年同月增減(%)':'去年同月增減'}, inplace=True)
df.rename(columns={'累計營業收入-當月累計營收':'當月累計營收'}, inplace=True)
df.rename(columns={'累計營業收入-去年累計營收':'去年累計營收'}, inplace=True)
df.rename(columns={'累計營業收入-前期比較增減(%)':'前期比較增減'}, inplace=True)

df.head(5)
```

```python
# 設定索引

df.set_index(['資料年月','公司代號'], inplace=True)
df.head(5)
```

# 寫入 csv 檔案

```python
filename = '營業收入-'+ year + month +'.csv'
df.to_csv(filename, encoding='utf-8-sig')
print('寫入完畢!')
```

# SQLite 資料庫

SQLite：一種檔案型式的資料庫，[官方網站](https://www.sqlite.org/index.html)

DB Browser for SQLite：讀取及操作 SQLite 工具軟體，[官方網站](https://sqlitebrowser.org/)

Python 操作 SQLite 教學：[英文教學網站](https://sebastianraschka.com/Articles/2014_sqlite_in_python_tutorial.html)

```python
import sqlite3

# 建立資料庫連結
conn = sqlite3.connect('stock.db')  # or stock.sqlite3

# 寫入資料庫
df.to_sql('monthly', conn, if_exists='replace')

print('寫入完畢!')
```
