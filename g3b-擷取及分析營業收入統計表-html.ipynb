# 月營收資料

* 至 [公開資訊觀測站：彙總報表/營運概況/每月營收/採用IFRSs後每月營業收入彙總表](https://mops.twse.com.tw/mops/web/t21sc04_ifrs)頁面
* 輸入營業收入年度及月份
* 找到 html 檔案的下載網址 (https://mops.twse.com.tw/nas/t21/sii/t21sc03_年_月_0.html)
* 使用 pandas 套件下載及讀取資料
* 將資料經適當的處理 (刪除資料、條件式過濾...等)
* 儲存資料

```python
import pandas as pd
import requests, sqlite3
```

```python
# 輸入月營收年度及月份

year = input('請輸入民國年度（譬如:109）：')
month = input('請輸入月份（譬如:2）：')
```

```python
# 從 html 檔擷取資料

try:
    url = 'https://mops.twse.com.tw/nas/t21/sii/t21sc03_'+year+'_'+month+'_0.html'
    data = requests.get(url)
    #dfs = pd.read_html(url, encoding='big5')
    #dfs = pd.read_html('test1.html', encoding='utf-8')
except Exception as e:
    print('**WARRN: 擷取失敗')
    print('錯誤訊息', e)
else:
    print('成功擷取!')
```

```python
# 設定字串編碼

#data.encoding = 'big5'
data.text[:1000]
```

```python
# 讀取資料到 dataframe

dfs = pd.read_html(data.text)
```

```python
# 讀取 dataframe 資料

print(len(dfs))   # 共有幾個表格
dfs[24]
```

```python
# 查看 dataframe 資料，每個表格的 shape 即幾列幾欄，前為列後為欄

#dfs[5].shape

for i in range(len(dfs)):
    print(i, dfs[i].shape)
```

```python
# 欄位數為 11 的,才是我們要保留的

newdf = pd.DataFrame()

for df in dfs:
    if df.shape[1] == 11:
        newdf = pd.concat([newdf, df])

# 上述 for 迴圈可簡寫為下列一行程式碼
# newdf = pd.concat([newdf for newdf in dfs if newdf.shape[1] == 11])
        
newdf
```

```python
# 設定 dataframe 欄位名稱

newdf.columns
#newdf.columns.get_level_values(1)
#newdf.columns = newdf.columns.get_level_values(1)
```

```python
# 移除不需要的資料：公司代號為合計者

#newdf['公司代號'] == '合計'
#newdf[newdf['公司代號'] == '合計']
newdf = newdf[newdf['公司代號'] != '合計']
newdf.tail(5)
```

```python
# 新增日期欄位

date = year + '/' + month
newdf.loc[:, ('date')] = date
newdf.head(5)
```

```python
# 設定索引值

newdf = newdf.set_index(['date', '公司代號'])
newdf.head(5)
```

# 寫入及讀取 csv 檔案

```python
# 寫入 csv 檔

newdf.to_csv('montyly.csv', encoding='utf-8-sig')
print('Save ok!')
```

```python
# 讀取 csv 檔

df = pd.read_csv('montyly.csv', index_col=['date', '公司代號'])
df.head(5)
```

# 寫入 SQLite

```python
import sqlite3

conn = sqlite3.connect('stock.sqlite3')
newdf.to_sql('monthly', conn, if_exists='replace')
print('Save ok!')
```

```python
# 讀取資料庫

df = pd.read_sql('select * from monthly', conn, index_col='公司代號')
df.tail(5)
```
