# 使用 requests 套件擷取html資料

```python
import requests
```

```python
# 讀取網頁資料

url = 'https://tw.yahoo.com/'
data = requests.get(url)

if data.status_code == requests.codes.ok:
    data.encoding = 'utf-8'
    print(data.text)

f = open('page.html', 'w', encoding='utf-8')
f.write(data.text)
f.close
```
