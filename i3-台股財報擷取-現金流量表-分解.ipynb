# 台股財報擷取-現金流量表

網址：[公開資訊觀測站/財務報表/採IFRS後/合併/個別報表/現金流量表](https://mops.twse.com.tw/mops/web/t164sb05)

```python
import requests, os, sqlite3
import pandas as pd
```

```python
print('擷取台股財報－現金流量表...')
sid = input('請輸入個股代號：')
year = input('請輸入擷取年度（請輸入民國年，譬如：108）：')
season = input('請輸入擷取季（輸入格式及區間：01 ~ 04）：')
```

```python
# 設定網址及 post 表單參數

url = 'https://mops.twse.com.tw/mops/web/t164sb05'
form_data = {
    'encodeURIComponent':1,
    'step':1,
    'firstin':1,
    'off':1,
    'queryName':'co_id',
    'inputType':'co_id',
    'TYPEK':'all',
    'isnew':'false',
    'co_id':sid,
    'year':year,
    'season':season,
}
```

```python
# 擷取資料

try:
    data = requests.post(url, data=form_data)
except Exception as e:
    print('**WARRN: 擷取失敗')
    print('錯誤訊息', e)
else:
    print('成功擷取！')
    #print(data.text[:1000])
```

```python
# 使用 pandas 讀取網頁裡的表格資料

dfs = pd.read_html(data.text)
#dfs = pd.read_html('i3-cashflow.html')
```

```python
# 取得現金流量表表格資料

# print(len(dfs))
# dfs[10]
df = dfs[10]
df.head(5)
```

```python
# 查看列數及欄數

df.shape
```

```python
# 重設欄位名稱

df.columns = list(range(df.shape[1]))
df.head(5)
```

```python
# 設定索引值

df = df.set_index(0)
df.head(5)
```

```python
# 只取第一欄資料(即當季財報), 及刪除 nan 項

df1 = df[1].dropna()
df1
```

```python
# 只取1個欄位資料後, bsdf 變成 series, 需轉換為 Dataframe

df1 = pd.DataFrame(df1)
```

```python
# 轉置表格, 即列與欄交換

df1 = df1.transpose()
df1
```

```python
# 刪除重覆欄位

df1 = df1.loc[:, ~df1.columns.duplicated()]
df1
```

```python
# 新增股票代碼及財報日期欄位

df1['sid'] = sid
df1['year'] = year
df1['season'] = season

# 設定索引值

df1 = df1.set_index(['sid', 'year', 'season'])
df1
```

```python
conn = sqlite3.connect(os.path.join('data', 'stock1.db'))

# 判斷 income table 是否已存在
query = "select count(*) from sqlite_master where type='table' and name='cashflow'"
exist = list(conn.execute(query))[0][0]

# 如果 bs table 存在，新增資料，否則，建立表格及新增資料
if exist:
    # 讀取資料庫中原有資料
    data = pd.read_sql('select * from cashflow', conn, index_col=(['sid', 'year', 'season']))
    
    # 合併原有資料與新資料
    newdf = pd.concat([data, df1], sort=True)

    # 刪除重覆項
    newdf.drop_duplicates(inplace=True)
    
    newdf.to_sql('cashflow', conn, if_exists='replace')
    print('資料新增完畢!')
else:
    df1.to_sql('cashflow', conn, if_exists='replace')
    print('資料儲存完畢!')

conn.close()
```
